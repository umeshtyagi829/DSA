```python
class CircularLinkedList:
    def __init__(self, data):
        self.data = data
        self.nextNode = None
        
    def NextNode(self, nextnode):
        self.nextNode = nextnode
```

```python
item1 = CircularLinkedList("a@gmail.com")
item2 = CircularLinkedList("b@gmail.com")
item3 = CircularLinkedList("c@gmail.com")
item4 = CircularLinkedList("d@gmail.com")
item5 = CircularLinkedList("e@gmail.com")
```

```python
nodeDetails = [item1.__dict__,
item2.__dict__,
item3.__dict__,
item4.__dict__,
item5.__dict__]
```
